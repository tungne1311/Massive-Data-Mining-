## Cấu hình

```python
import os
import warnings
import gc
warnings.filterwarnings('ignore')

import pandas as pd
import pyarrow as pa
from huggingface_hub import hf_hub_download, snapshot_download

CATEGORY   = "Electronics"
HF_REPO    = "McAuley-Lab/Amazon-Reviews-2023"
OUTPUT_DIR = "/kaggle/working/amazon_electronics"
os.makedirs(OUTPUT_DIR, exist_ok=True)

import psutil
ram_gb = psutil.virtual_memory().total / 1024**3
print(f"RAM khả dụng: {ram_gb:.0f} GB")
```

## Tải Review Data

```python
print(f"Đang tải Review data cho '{CATEGORY}' (Pandas)...")

review_local_path = hf_hub_download(
    repo_id=HF_REPO,
    repo_type="dataset",
    filename=f"raw/review_categories/{CATEGORY}.jsonl"
)

# Đọc file bằng Pandas
review_df = pd.read_json(review_local_path, lines=True)

print(f"Review records : {len(review_df):,}")
print(f"Columns        : {list(review_df.columns)}")
display(review_df.head(3))
```

## Tải Meta Data

```python
print(f"Đang tải Meta data cho '{CATEGORY}' (Pandas)...")

meta_folder_path = snapshot_download(
    repo_id=HF_REPO,
    repo_type="dataset",
    allow_patterns=f"raw_meta_{CATEGORY}/*.parquet"
)

import glob
parquet_files = glob.glob(os.path.join(meta_folder_path, f"raw_meta_{CATEGORY}/*.parquet"))
meta_df = pd.concat([pd.read_parquet(f) for f in parquet_files], ignore_index=True)

print(f"Meta records   : {len(meta_df):,}")
print(f"Columns        : {list(meta_df.columns)}")
display(meta_df.head(3))
```

## Lọc dữ liệu

```python
print("Đang lọc các cột cần thiết...")

review_cols = ['parent_asin', 'rating', 'timestamp', 'user_id']
review_df = review_df[review_cols]

print("Chỉ giữ lại các review có rating >= 4...")
review_df = review_df[review_df['rating'] >= 4]

meta_cols = ['title', 'rating_number', 'parent_asin', 'details']
meta_df = meta_df[meta_cols]

print("--- Review Data Info ---")
review_df.info()

print("\n--- Meta Data Info ---")
meta_df.info()
```

```python
before = len(review_df)
print(f"Trước dedup: {before:,} reviews")

# Với mỗi cặp (user_id, parent_asin), chỉ giữ review mới nhất theo timestamp
review_df = (
    review_df
    .sort_values("timestamp", ascending=False)
    .drop_duplicates(subset=["user_id", "parent_asin"], keep="first")
    .sort_values("timestamp")
    .reset_index(drop=True)
)

after = len(review_df)
print(f"Sau dedup : {after:,} reviews")
print(f"Đã xóa    : {before - after:,} duplicate reviews")
```

## Lưu dữ liệu

```python
import os

drive_path = "/kaggle/working/Data_Amazon_Electronics"
os.makedirs(drive_path, exist_ok=True)

print("Đang lưu Review Data vào ổ đĩa...")
review_df.to_parquet(f"{drive_path}/review_data.parquet", index=False)

print("\nĐang lưu Meta Data vào ổ đĩa...")
meta_df.to_parquet(f"{drive_path}/meta_data.parquet", index=False)

print(f"\nĐã lưu thành công dữ liệu vào: {drive_path}")
```

## Lọc dữ liệu K-core (k=5)

```python
k = 5
iteration = 1
while True:
    start_len = len(review_df)
    
    # Lọc user có >= k đánh giá
    user_counts = review_df['user_id'].value_counts()
    valid_users = user_counts[user_counts >= k].index
    review_df = review_df[review_df['user_id'].isin(valid_users)]
    
    # Lọc item (parent_asin) có >= k đánh giá
    item_counts = review_df['parent_asin'].value_counts()
    valid_items = item_counts[item_counts >= k].index
    review_df = review_df[review_df['parent_asin'].isin(valid_items)]
    
    end_len = len(review_df)
    print(f"Lần lặp {iteration}: Số lượng review giảm từ {start_len:,} -> {end_len:,}")
    
    if start_len == end_len:
        print("Đã hội tụ!")
        break
    iteration += 1

print(f"\nSố lượng review cuối cùng sau khi lọc 5-core: {len(review_df):,}")
print(f"Số lượng Users: {review_df['user_id'].nunique():,}")
print(f"Số lượng Items: {review_df['parent_asin'].nunique():,}")

print("\nĐang cập nhật lại Meta Data...")
valid_items_final = review_df['parent_asin'].unique()
meta_df = meta_df[meta_df['parent_asin'].isin(valid_items_final)].reset_index(drop=True)
print(f"Số lượng Meta records cuối cùng: {len(meta_df):,}")
```

```python
print("Đang lưu Review Data vào ổ đĩa...")
review_df.to_parquet(f"{drive_path}/review_data_5_core.parquet", index=False)

print("\nĐang lưu Meta Data vào ổ đĩa...")
meta_df.to_parquet(f"{drive_path}/meta_data_5_core.parquet", index=False)

print(f"\nĐã lưu thành công dữ liệu vào: {drive_path}")
```
